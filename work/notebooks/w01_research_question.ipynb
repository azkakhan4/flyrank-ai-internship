## 3. Quick look at the data

The starter dataset contains **30,000 rows** and **44 columns**, providing enough information to explore search performance patterns. The trend data also shows that **16,262 pages are trending down**, which suggests there are many pages that may benefit from content optimization. These numbers indicate that the dataset is large enough and contains useful information for building a machine learning solution.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/azkakhan4/flyrank-ai-internship/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

print("Number of rows:", len(df))
print("Number of columns:", len(df.columns))

print("\nAverage CTR:", round(df["ctr"].mean(), 2))
print("Average Position:", round(df["avg_position"].mean(), 2))

print("\nTrend Direction Counts:")
print(df["trend_direction"].value_counts())
```
